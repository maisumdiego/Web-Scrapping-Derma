## Definição da funções de inicialização

```python
def login(username, password):
    select_username = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='username']")))
    select_password = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "input[name='password']")))

    select_username.send_keys(username)
    select_password.send_keys(password)

    select_username.clear()
    select_password.clear()
    username = ''
    password = ''

    submit = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.CSS_SELECTOR, "button[type='submit']")))
    submit.click()
```

```python
def agora_nao():
    not_now = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Agora não')]")))
    not_now.click()

    not_now2 = WebDriverWait(browser, 10).until(
    EC.element_to_be_clickable((By.XPATH, "//button[contains(text(), 'Agora não')]")))
    not_now2.click()
```

## Definição de funções de raspagem de dados

```python
def busca_hashtag(tag):
    url = 'https://www.instagram.com/tags/' + tag
    browser.get(url)
```

```python
def coleta_links():
    num_posts = 0
    posts = []
    while num_posts < 210:
        browser.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        links = browser.find_elements(By.XPATH, "//a")
        sleep(1)

        for link in links:
            try:
                post = link.get_attribute('href')
                if '/p/' in post and post not in posts:
                    posts.append(post)
                    num_posts = num_posts + 1
            except:
                pass
    
    return posts
```

```python
class Operacoes():
    def usuario():
        try:
            user1 = browser.find_elements(By.XPATH, "//article/div/div[2]/div/div[1]/div/header/div[2]/div[1]/div[1]/div/div/div[1]/div/span/a")
            username = user1[0].get_attribute('innerText')
        except:
            user1 = browser.find_elements(By.XPATH, "//article/div/div[2]/div/div[1]/div/header/div[2]/div[1]/div/div/div/div[1]/a")
            user2 = browser.find_elements(By.XPATH, "//article/div/div[2]/div/div[1]/div/header/div[2]/div[1]/div/div/div/div[3]/a")
            username = [user1[0].get_attribute('innerText'), user2[0].get_attribute('innerText')]
        return username

    def data():
        try:
            local_data = browser.find_elements(By.XPATH, "//main/div[1]/div[1]/article/div/div[2]/div/div[2]/div[2]/div/div/a/div/time")
            data_original = local_data[0].get_attribute('datetime')
            data_original = data_original[0:10] + '-' + data_original[11:-5]
            date = datetime.strptime(data_original, '%Y-%m-%d-%H:%M:%S')
        except:
            date = None
        return date

    def legenda():
        try:
            local_desc = browser.find_elements(By.XPATH, "//article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/div[1]/h1")
            desc = local_desc[0].get_attribute('innerText')
        except:
            desc = None
        return desc

    def curtidas():
        local_likes = browser.find_elements(By.XPATH, "//div[1]/div[1]/article/div/div[2]/div/div[2]/section[2]/div/div/div/a/div/span")
        try:
            likes = local_likes[0].get_attribute('innerText')
        except:
            likes = None
        return likes

    def nome():
        try:
            local_name = browser.find_elements(By.XPATH, "//div[1]/div[2]/section/main/div/header/section/div[3]/span")
            name = local_name[0].get_attribute('innerText')
        except:
            name = None
        return name

    def total_post():
        try:
            local_posts = browser.find_elements(By.XPATH, "//section/main/div/header/section/ul/li[1]/div/span/span")
            posts = local_posts[0].get_attribute('innerText')
        except:
            posts = None
        return posts

    def biografia():
        try:
            local_bio = browser.find_elements(By.XPATH, "//section/main/div/header/section/div[3]/h1")
            bio = local_bio[0].get_attribute('innerText')
        except:
            bio = None
        return bio

    def seguidores():
        try:
            local_num_followers = browser.find_elements(By.XPATH, "//section/main/div/header/section/ul/li[2]/a/div/span")
            num_followers = local_num_followers[0].get_attribute('title')
            if num_followers == '': num_followers = None
        except:
            num_followers = None
        return num_followers

    def seguindo():
        try:
            local_num_following = browser.find_elements(By.XPATH, "//section/main/div/header/section/ul/li[3]/a/div/span/span")
            num_following = local_num_following[0].get_attribute('innerText')
        except:
            num_following = None
        return num_following
```

```python
def raspagem(links):
    erros = 0
    lista_id = []
    lista_date = []
    lista_username = []
    lista_desc = []
    lista_likes = []
    lista_url = []
    lista_name = []
    lista_posts = []
    lista_bio = []
    lista_num_followers = []
    lista_num_following = []
    lista_data_coleta = []
    
    for i, post in enumerate(links):
        try:
            browser.get(post)
            sleep(3)

            # RASPAGEM POSTAGEM
                # MONTAGEM DE LISTA
            usuario_amostra = Operacoes.usuario()
            lista_date.append(Operacoes.data())
            lista_username.append(usuario_amostra)
            lista_desc.append(Operacoes.legenda())
            lista_likes.append(Operacoes.curtidas())
            lista_url.append(post)
            lista_data_coleta.append(datetime.today())
          

            # RASPAGEM DO PERFIL
                # CLICANDO NO PERFIL
            try:
                perfil = WebDriverWait(browser, 0).until(EC.element_to_be_clickable((By.XPATH, "//header/div[2]/div[1]/div[1]/div/div/div[1]/div/span/a")))
                perfil.click()
            except:
                perfil = WebDriverWait(browser, 10).until(EC.element_to_be_clickable((By.XPATH, "//article/div/div[2]/div/div[2]/div[1]/ul/div/li/div/div/div[2]/h2/div/div/a")))
                perfil.click()
            sleep(3)
            
                # MONTANDO DE LISTAS
            nome_amostra = Operacoes.nome()
            lista_name.append(nome_amostra)
            lista_posts.append(Operacoes.total_post())
            lista_bio.append(Operacoes.biografia())
            lista_num_followers.append(Operacoes.seguidores())
            lista_num_following.append(Operacoes.seguindo())
          
            print(f'{i}. OK')
            erros = 0

        except Exception as e:
            print(f'{i}. A postagem do endereço {post} apresentou erro: "{e}"')
            erros = erros + 1
            pass
        
        #if erros == 3: break

    uniao_listas = lista_username, lista_likes, lista_desc, lista_date, lista_data_coleta, lista_url, lista_name, lista_posts, lista_num_followers, lista_num_following, lista_bio

    return uniao_listas
```

## Definição de funções de tratamento de dados

```python
def tabela(listas):
    dados = pd.DataFrame()
    colunas = ['usuario', 'curtidas', 'legenda', 'data_origem', 'coleta', 'url', 'nome', 'total_posts', 'seguidores', 'seguindo', 'bio']
    for i in range(len(listas)):
        dados[f'{colunas[i]}'] = lista[i]
    return dados
```

```python
class Ajustes():
    def numeros(dataset):
        colunas_numerais = ['seguidores', 'seguindo', 'curtidas', 'total_posts']
        for i in colunas_numerais:
            dataset[i] = dataset[i].fillna('0')
            dataset[i] = dataset[i].apply(lambda x: x.replace('.', ''))
            dataset[i] = dataset[i].astype('int')

    def data(dataset):
        dataset['data_ajustada'] = dataset.data_origem.dt.tz_localize(0).dt.tz_convert('America/Sao_Paulo')
        dataset['data_ajustada'] = dataset['data_ajustada'].dt.tz_localize(None)
        dataset['data_postagem'] = dataset.data_ajustada.apply(lambda x: x.date())
        dataset['horario_postagem'] = dataset.data_ajustada.apply(lambda x: x.time())

    def data_coleta(dataset):
        dataset['data_coleta'] = dataset['coleta'].apply(lambda x: x.date())
        dataset['horario_coleta'] = dataset['coleta'].apply(lambda x: x.time())
    
    def coluna_hashtag(dataset):
        lista_hashtags = []
        for i in range(len(dataset)):
            lista_intermediaria = ''
            if dataset['legenda'][i] is None:
                lista_intermediaria = None
            else:
                for palavra in dataset['legenda'][i].split():
                    if palavra[0] == '#': 
                        lista_intermediaria = lista_intermediaria + ' ' + palavra
            lista_hashtags.append(lista_intermediaria)
        dataset['tags'] = lista_hashtags
```
